# Gray Scale to RGB

```python
#Read a Gray Scale image
image=cv2.imread('test (1).jpg')
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Convert an image to tensor
tensor_image=kornia.utils.image_to_tensor(img,keepdim=False)

#Convert a Gray Scale to RGB image
img1=kornia.color.grayscale_to_rgb(tensor_image)

#Convert a tensor to image
img1= kornia.tensor_to_image(img1)

#Subplotting the images
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img,cmap='gray')
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(img1)
```

# BGR to RGB

```python
#Read a BGR image
image=cv2.imread("test (1).jpg")
#img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Convert an image to tensor
tensor_image=kornia.utils.image_to_tensor(image, keepdim=False).float()

#Convert an RGB to BGR image
rgb=kornia.color.bgr_to_rgb(tensor_image)

#Convert a tensor to image
img1= kornia.tensor_to_image(rgb.byte())

#Subplotting the images
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(image)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(img1)
```

# RGB to Gray

```python
#Read an RGB image
image=cv2.imread("test (1).jpg")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Convert an image to tensor
tensor_image=kornia.utils.image_to_tensor(img, keepdim=False).float()

#Convert an RGB to Gray Scale image
gray=kornia.color.rgb_to_grayscale(tensor_image)

#Convert a tensor to image
img1= kornia.tensor_to_image(gray)

#Save the gray scale image
cv2.imwrite('rTOg.jpg',img1)

#Subplotting the images
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(img1,cmap="gray")
```


# RGB to HLS

```python
#Read an RGB image
image=cv2.imread("test (1).jpg")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Convert an image to tensor
tensor_image=kornia.utils.image_to_tensor(img, keepdim=False).float()

#Convert an RGB to HLS image
hls=kornia.color.rgb_to_hls(tensor_image)

#Convert a tensor to image
img1= kornia.tensor_to_image(hls)

#Save the gray scale image
cv2.imwrite('rTOhlc.png',img1)

#Subplotting the images
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(img1)
```

# RGB to HSV

```python
#Read an RGB image
image=cv2.imread("test (1).jpg")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Convert an image to tensor
tensor_image=kornia.utils.image_to_tensor(img, keepdim=False).float()

#Convert an RGB to HSV image
h=kornia.color.rgb_to_hsv(tensor_image)

#Convert a tensor to image
img1= kornia.tensor_to_image(h)

#Save the gray scale image
cv2.imwrite('rTOhsv.png',img1)

#Subplotting the images
fig = plt.figure()
ax1 = fig.add_subplot(1,5,1)#subplot(nrows, ncols, index, **kwargs)#subplot(pos, **kwargs)#subplot(**kwargs)#subplot(ax)
ax1.imshow(img)
ax2 = fig.add_subplot(1,5,5)
ax2.imshow(img1)
```
